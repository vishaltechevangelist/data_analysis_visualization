```python
# plot the linear graph
import matplotlib.pyplot as plt
x = [0,1,2,3,4]
y = [0,1,4,9,16]
plt.plot(x,y)
plt.show()
```

```python
x = [0,2,4,6,8]
y = [0,4,16,32,64]
fig, ax = plt.subplots()

ax.plot(x,y,marker = 'o', label = 'data Points')
ax.set_title('This is first plot')
ax.set_xlabel('Date')
ax.set_ylabel('Performance')
plt.show()
```

```python
# combine line chart in same plot
import numpy as np
x = np.array([1,2,3,4])
y = x*2
plt.plot(x,y)

x1 = [2,4,6,8]
y1 = [3,5,7,9]
plt.plot(x1, y1, '-.') # plot dotted line
plt.xlabel("Date")
plt.ylabel("Performance")
plt.title('My first combine plot')
plt.show()
```
